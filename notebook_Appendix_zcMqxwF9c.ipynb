```python
import matplotlib.pyplot as plt

# Data
data = [
    ['Los Angeles', 640],
    ['New York', 3226],
    ['San Francisco', 435],
    ['Washington DC', 5316],
    ['Seattle', 3375],
    ['Austin', 434],
    ['Detroit', 3945]
]

# Sort the data by descending order of number of jobs
data.sort(key=lambda x: x[1], reverse=True)
locations = [item[0] for item in data]
number_of_jobs = [item[1] for item in data]

# Different colors for bars
colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'cyan']

# Create a bar graph
plt.figure(figsize=(10, 6))
bars = plt.bar(locations, number_of_jobs, color=colors)
plt.xlabel('Location')
plt.ylabel('Number of Jobs')
plt.title('Number of Jobs by Location (Sorted by Descending)')
plt.xticks(rotation=45)
plt.tight_layout()
for bar, value in zip(bars, number_of_jobs):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(value), ha='center', va='bottom')

# Display the graph
plt.show()
```

```python
data = [
    ('C', 13498),
    ('C#', 333),
    ('C++', 305),
    ('Java', 2609),
    ('JavaScript', 355),
    ('Python', 1173),
    ('Scala', 33),
    ('Oracle', 784),
    ('SQL Server', 250),
    ('MySQL Server', 0),
    ('PostgreSQL', 10),
    ('MongoDB', 174)
]

# Sort the data by descending order of number of jobs
data.sort(key=lambda x: x[1], reverse=True)
languages = [item[0] for item in data]
number_of_jobs = [item[1] for item in data]

# Define custom colors for bars
colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan', 'pink', 'gray', 'magenta', 'gold', 'skyblue', 'lime']

# Create a bar graph
plt.figure(figsize=(10, 6))
bars = plt.bar(languages, number_of_jobs, color=colors)
plt.xlabel('Programming Languages')
plt.ylabel('Number of Jobs')
plt.title('Number of Jobs by Programming Language (Sorted by Descending)')
plt.xticks(rotation=45)
plt.tight_layout()

# Add value annotations on top of the bars
for bar, value in zip(bars, number_of_jobs):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(value), ha='center', va='bottom')

# Display the graph
plt.show()
```
